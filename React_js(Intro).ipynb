# **what is mean by ReactJS ?**

**ReactJS is a JavaScript library used to build user interfaces, especially for web applications. It allows developers to create reusable components (like buttons, forms, etc.) that make up a webpage. React helps update only the parts of the page that change, making the app faster.**

**Free and Open Source:** ReactJS is free to use, with its code open for anyone to contribute or use.

**Developed by Facebook:** React was created by Facebook (Meta) and is maintained by them and a global community.

**Single Page Applications (SPA):** React is ideal for building SPAs, where content updates without reloading the entire page.

**Web and Mobile Applications:** React builds dynamic web apps, while React Native lets you build mobile apps for iOS and Android using the same codebase.

# **Why it is used ?**

**Component-Based Architecture:** React organizes the UI into reusable components, making development modular and efficient.

Ex: In a React application, different parts of a website, like Header, Navbar, Content, or Product, can be treated as separate components.

**Uses Virtual DOM for Fast Rendering:** React employs a virtual DOM, which updates only the necessary parts of the UI, enhancing performance.

**Easy to Learn:** React’s simplicity and strong community support make it beginner-friendly.

# **Explanation of the React & ReactDOM CDN Scripts**

**These script tags load React and ReactDOM from a CDN (Content Delivery Network), allowing you to use React in your HTML file without installing it locally.**

# **1️⃣ React Library (react.development.js)**



```
<script crossorigin src="https://unpkg.com/react@18/umd/react.development.js"></script>

```



🔹 Loads the core React library (React API, components, state management, hooks, etc.).

🔹 Used to create UI components and manage state in React applications.

🔹 Development version → includes extra debugging features & warnings for easier debugging.



# **2️⃣ ReactDOM Library (react-dom.development.js)**



```
<script crossorigin src="https://unpkg.com/react-dom@18/umd/react-dom.development.js"></script>

```



🔹 Provides the ReactDOM API to render React components into the browser's DOM.

🔹 Required for rendering React elements in index.html (like ReactDOM.createRoot(...)).

🔹 Development version → includes helpful warnings & debugging tools.
