# **What is React.memo?**

React.memo is a higher-order component (HOC) in React that helps optimize performance by preventing unnecessary re-renders of a component.

When to Use It?
Use React.memo when:
âœ… A component receives the same props frequently.
âœ… The component doesnâ€™t need to re-render unless its props change.

# **How it helps?**

Without React.memo: The component re-renders every time the parent updates, even if the props donâ€™t change.

With React.memo: The component only re-renders when the props actually change, improving performance.

# **what is mean by  custom hook?**

**A custom hook in React is a reusable function that uses built-in hooks (useState, useEffect, etc.) to add functionality to multiple components without repeating code.**

**Example**

ðŸ‘‰ Instead of writing the same logic in multiple components, we create a custom hook and reuse it.

# **Create PageTitleOne Component**



```
import React, { useState } from "react";
import usePageTitle from "./usePageTitle";

const PageTitleOne = () => {
  const [count, setCount] = useState(0);

  usePageTitle(count);

  return (
    <div>
      <h2>Page Title One</h2>
      <button onClick={() => setCount(count + 1)}>Count - {count}</button>
    </div>
  );
};

export default PageTitleOne;

```



# **Create PageTitleTwo Component**



```
import React, { useState } from "react";
import usePageTitle from "./usePageTitle";

const PageTitleTwo = () => {
  const [count, setCount] = useState(0);

  usePageTitle(count);

  return (
    <div>
      <h2>Page Title Two</h2>
      <button onClick={() => setCount(count + 1)}>Count - {count}</button>
    </div>
  );
};

export default PageTitleTwo;

```



# **usePageTitle.js**



```
import { useEffect } from "react";

function usePageTitle(count) {
  useEffect(() => {
    document.title = `Count - ${count}`;
  }, [count]);
}

export default usePageTitle;

```



# **App.js**



```
import React from "react";
import PageTitleOne from "./PageTitleOne";
import PageTitleTwo from "./PageTitleTwo";

const App = () => {
  return (
    <div>
      <PageTitleOne />
      <PageTitleTwo />
    </div>
  );
};

export default App;

```
