# **What are Hooks in React JS?**
Hooks are special functions in React JS that let you use state and other React features in functional components (without needing class components).

# **Why Use Hooks?**
Before Hooks, you had to use class components for features like state and lifecycle methods. Hooks make it easier and cleaner to use these features in functional components.

# **Example 1: useState Hook (Manages State)**

useState lets you add state to a functional component.



```
import { useState } from "react";

function Counter() {
  const [count, setCount] = useState(0); // useState Hook

  return (
    <div>
      <p>Count: {count}</p>
      <button onClick={() => setCount(count + 1)}>Increase</button>
    </div>
  );
}

export default Counter;

```



# **How it works?**

useState(0) initializes state (count) with 0.

setCount(count + 1) updates the count when the button is clicked.

# **Example 2: useEffect Hook (Handles Side Effects)**

useEffect runs after the component renders. It is used for fetching data,
subscriptions, or updating the DOM.



```
import { useState, useEffect } from "react";

function Timer() {
  const [seconds, setSeconds] = useState(0);

  useEffect(() => {
    const interval = setInterval(() => {
      setSeconds((prev) => prev + 1);
    }, 1000);

    return () => clearInterval(interval); // Cleanup
  }, []); // Empty dependency array = Runs only once

  return <p>Timer: {seconds}s</p>;
}

export default Timer;

```



# **useContext – Shares Data Between Components**

Avoids prop drilling by providing a global state.



```
import React, { useContext, createContext } from "react";

const UserContext = createContext(); // Create a context

function Child() {
  const user = useContext(UserContext); // Access context value
  return <h2>Welcome, {user}!</h2>;
}

function Parent() {
  return (
    <UserContext.Provider value="Alice">
      <Child />
    </UserContext.Provider>
  );
}

export default Parent;

```



✅ Use: Theme, user authentication, global settings.

# **Additional Hooks (For Advanced Use Cases)**

🔹 useRef – Access DOM Elements & Store Values Without Re-rendering
Keeps a value without causing re-renders.



```
import React, { useRef } from "react";

function InputFocus() {
  const inputRef = useRef(null); // Create a reference

  return (
    <div>
      <input ref={inputRef} type="text" />
      <button onClick={() => inputRef.current.focus()}>Focus Input</button>
    </div>
  );
}

export default InputFocus;

```



✅ Use: Manipulating DOM elements, storing previous values.

# **🔹 useReducer – Alternative to useState for Complex Logic**
Works like Redux but inside components.



```
import React, { useReducer } from "react";

const reducer = (state, action) => {
  if (action.type === "INCREMENT") return { count: state.count + 1 };
  return state;
};

function Counter() {
  const [state, dispatch] = useReducer(reducer, { count: 0 });

  return (
    <div>
      <p>Count: {state.count}</p>
      <button onClick={() => dispatch({ type: "INCREMENT" })}>Increase</button>
    </div>
  );
}

export default Counter;

```



✅ Use: When state logic is complex (e.g., forms, shopping cart).

# **useMemo – Optimizes Performance**
Caches expensive calculations to avoid unnecessary re-computations.



```
import React, { useState, useMemo } from "react";

function ExpensiveCalculation({ number }) {
  const result = useMemo(() => {
    console.log("Calculating...");
    return number * 2;
  }, [number]);

  return <p>Result: {result}</p>;
}

export default function App() {
  const [count, setCount] = useState(1);

  return (
    <div>
      <ExpensiveCalculation number={count} />
      <button onClick={() => setCount(count + 1)}>Increase</button>
    </div>
  );
}

```



✅ Use: Speeds up slow functions by storing results.

**🔹 useCallback – Prevents Unnecessary Function Re-Creation**

Memoizes functions to avoid unnecessary re-renders.



```
import React, { useState, useCallback } from "react";

function Button({ onClick }) {
  console.log("Button re-rendered");
  return <button onClick={onClick}>Click Me</button>;
}

export default function App() {
  const [count, setCount] = useState(0);

  const handleClick = useCallback(() => {
    setCount((prev) => prev + 1);
  }, []);

  return (
    <div>
      <p>Count: {count}</p>
      <Button onClick={handleClick} />
    </div>
  );
}

```



✅ Use: Prevents unnecessary re-rendering of components.

# **Other Common Hooks**

useContext – Access global values without passing props.

useRef – Get a reference to a DOM element or store values without re-rendering.

useReducer – An alternative to useState for complex state logic.

useMemo & useCallback – Optimize performance by memoizing values and functions.
