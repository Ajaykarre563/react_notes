# **By using OBJECT**



```
import React, { useState } from "react"; // Importing React and useState hook

const Index = () => {
  // Initial state object with firstName, lastName, and age
  //const initialObj = {
    //firstName: "Ajay",
    //lastName: "Karre",
    //age: 23, // Fixed missing comma and removed incorrect assignment syntax

    // Function that returns the initial state object
  function initialObj() {
    return {
      firstName: "Ajay",
      lastName: "Karre",
      age: 23,
    };
  }
  // useState hook to manage the state
  const [data, setData] = useState(initialObj);

  // Function to change the state when the button is clicked
  const changeFirstName = () => {
    setData({
      ...data,         //only it change the first Name data and remaining data will be constant
      firstName: "Aryan", // Fixed object syntax
      //lastName: "Karthik",
      //age: 25, // Age should be a number, not a string
    });
  };

  const changelastName = () => {
    setData({
      ...data,         //only it change the last Name data and remaining data  will be constant
      //firstName: "Aryan", // Fixed object syntax
      lastName: "Karthik",
      //age: 25, // Age should be a number, not a string
    });
  };

  return (
    <div>
      <h1>Hello, my First name is {data.firstName}</h1>
      <button onClick={changeFirstName}>Change First Name</button> {/* Fixed onClick syntax */}
      <h1>Hello, my Last name is {data.lastName}</h1> {/* Fixed incorrect variable name */}
      <button onClick={changelastName}>Change last Name</button> {/* Fixed onClick syntax */}
      <h1>Age: {data.age}</h1>
    </div>
  );
};

export default Index;

```

# **By using ARRAY**



```
import React, { useState } from "react"; // Importing React and useState hook

const Index = () => {
  // ✅ Step 1: Define an initial array of objects
  const initialArray = [
    {
      id: "ak", // ✅ Fixed missing comma
      firstName: "Ajay",
      lastName: "Karre",
      age: 25,
    },
    {
      id: "sk", // ✅ Fixed missing comma
      firstName: "Aryan",
      lastName: "Aryan",
      age: 24,
    },
  ];

  // ✅ Step 2: Use the useState hook to manage the array state
  const [data, setData] = useState(initialArray);

  console.log(data); // Debugging: Logs the current state in the console

  // ✅ Step 3: Function to delete a user based on ID
  const deleteUser = (id) => {
    const updatedData = data.filter((eachItem) => eachItem.id !== id);
    setData(updatedData); // Update state with the filtered list
  };

  return (
    <div>
      <h1>Users List</h1>
      <ul>
        {data.map((eachItem) => {
          // Loop through each object in the array
          const { id, firstName, lastName, age } = eachItem;

          return (
            <li key={id}> {/* ✅ Using "id" as the unique key */}
              <div>First Name: {firstName}</div>
              <div>Last Name: {lastName}</div>
              <div>Age: {age}</div>
              {/* ✅ Step 4: Add delete button */}
              <button onClick={() => deleteUser(id)}>Delete</button>
            </li>
          );
        })}
      </ul>
    </div>
  );
};

export default Index;

```
