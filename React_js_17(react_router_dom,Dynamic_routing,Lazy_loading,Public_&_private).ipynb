# **what is mean by react-router-dom ?**

**react-router-dom is a library used in React applications for handling navigation and routing. It allows you to create single-page applications (SPAs) with multiple views without requiring a full page reload.**

**Key Features:**

**Declarative Routing:** Define routes using components.

**Dynamic Routing:** Routes change based on user interactions.

**Nested Routes**: Organize routes in a hierarchy.

**Navigation**: Use <Link> and <Navigate> instead of <a> to prevent page reloads.

# **Note:**

You can switch between different pages (or views) without changing the URL

However, if you use react-router-dom, the URL does change when navigating between pages.

So, if you want to view multiple pages without changing the URL, use state and conditional rendering instead of react-router-dom. âœ…



```
import { useState } from "react";

function Home() {
  return <h2>Home Page</h2>;
}

function About() {
  return <h2>About Page</h2>;
}

function App() {
  const [page, setPage] = useState("home");

  return (
    <div>
      <nav>
        <button onClick={() => setPage("home")}>Home</button>
        <button onClick={() => setPage("about")}>About</button>
      </nav>
      <div>
        {page === "home" && <Home />}
        {page === "about" && <About />}
      </div>
    </div>
  );
}

export default App;

```



**nav link is an HTML tag used to define a navigation section in a webpage. It is commonly used to hold links for navigating between different pages or sections of a website.**

**What is Navigate link in react-router-dom?**

Navigate link is a component in react-router-dom used to redirect users from one route to another automatically.

# **what is mean by Dynamic Routing in react-router-dom?**

Dynamic routing means creating routes that change based on the URL. Instead of making separate routes for each page, you use a placeholder in the URL to handle multiple pages with the same structure.

**Example**

Imagine you have a blog where each post has a unique ID. Instead of making separate routes like:

âœ… /post/1

âœ… /post/2

âœ… /post/3



```
<Route path="/post/:id" element={<Post />} />

```



**Here, :id is a placeholder that changes based on the URL.**

When a user visits /post/5, it shows Post 5.

When they visit /post/10, it shows Post 10.

# **what is mean by Lazy Loading in react-router-dom?**

**Lazy loading means loading a component only when it's needed, instead of loading everything at once. This helps improve performance by reducing the initial load time of your app.**

# **Why is it Used?**

Makes the app faster by loading only the required pages.
Reduces unnecessary loading of unused components.
Improves user experience, especially for large apps.



```
import Home from "./Home";
import About from "./About";

```





```
import { lazy, Suspense } from "react";
import { BrowserRouter as Router, Routes, Route } from "react-router-dom";

const Home = lazy(() => import("./Home"));  
const About = lazy(() => import("./About"));  

function App() {
  return (
    <Router>
      <Suspense fallback={<h2>Loading...</h2>}>
        <Routes>
          <Route path="/home" element={<Home />} />
          <Route path="/about" element={<About />} />
        </Routes>
      </Suspense>
    </Router>
  );
}

export default App;

```



**How It Works?**

lazy(() => import("./Home")) loads Home only when needed.(inspect=>network you can check performance).

Suspense tag shows "Loading..." while the component is being loaded.


# **what is mean by Public Routing ?**

Routes (or pages) that anyone can access without logging in.

Example: A website's homepage, about page, or contact page.



```
import { BrowserRouter as Router, Routes, Route } from "react-router-dom";
import Home from "./Home";

function App() {
  return (
    <Router>
      <Routes>
        <Route path="/" element={<Home />} /> {/* Public Route */}
      </Routes>
    </Router>
  );
}

export default App;

```



# **what is mean Private Routing: ?**

Routes that require authentication (login) before access.

Example: A dashboard page in a banking app where users must log in to view their account details.

**Example in a Web App**

**Imagine a shopping website:**

Public Route â†’ The homepage where anyone can see products.

Private Route â†’ The cart or order history page that only logged-in users can access.



```
import { Navigate } from "react-router-dom";

function PrivateRoute({ children }) {
  const isAuthenticated = localStorage.getItem("user"); // Example check
  return isAuthenticated ? children : <Navigate to="/" />;
}

export default PrivateRoute;

```



#**Example for public and private**



```
import { BrowserRouter as Router, Routes, Route } from "react-router-dom";
import Home from "./Home";
import Dashboard from "./Dashboard";
import PrivateRoute from "./PrivateRoute"; // Importing PrivateRoute

function App() {
  return (
    <Router>
      <Routes>
        <Route path="/" element={<Home />} /> {/* Public Route */}
        <Route
          path="/dashboard"
          element={
            <PrivateRoute>
              <Dashboard />
            </PrivateRoute>
          }
        /> {/* Private Route */}
      </Routes>
    </Router>
  );
}

export default App;

```
