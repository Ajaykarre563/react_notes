# **What is Debugging in React?**

Debugging in React means finding and fixing errors in your code. It helps you understand why something is not working in your React app.

# **Common Debugging Techniques in React**

Console Logs (console.log()) – Print values to check what's happening.

React Developer Tools – Inspect React components in the browser.

Error Boundaries – Catch and handle errors in components.

Debugger Statement – Pause code execution for deeper inspection.(line by line analysis)

Using React.StrictMode – Helps detect potential issues.

Example: Debugging with console.log()

**What is the "Network" Tab in the Console?**

The Network tab in your browser’s Developer Tools (DevTools) helps you monitor and debug network requests made by your React app.

It shows:
✅ API requests (fetch, Axios, etc.)

✅ Response status codes (200, 404, 500, etc.)

✅ Request/response time

✅ Headers, payload, and response data

**What is the "Components" Tab in React DevTools?**

The "Components" tab in React Developer Tools (React DevTools) helps you inspect, debug, and understand how React components work in your app.

🔍 With the Components tab, you can:

✅ See the component tree (parent-child relationships).

✅ View and edit props and state in real time.

✅ Find which component is causing errors.

✅ Check re-renders to optimize performance.

# **What is the "Profiler" Tab in React DevTools?**

The Profiler tab in React Developer Tools helps you analyze component performance by measuring how often components render and how long they take.

🔍 With the Profiler tab, you can:

✅ See which components re-render unnecessarily.

✅ Measure render time for each component.

✅ Find performance bottlenecks in your app.

✅ Optimize slow components to improve efficiency.
