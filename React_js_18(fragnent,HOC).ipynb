# **what is mean by  Fragment  in React js?**

**In React, a Fragment is a way to group multiple elements without adding an extra HTML tag like div tag.**

# **Why use Fragments?**

Avoid unnecessary div tag wrappers in the DOM.

Keep your HTML cleaner and more efficient.

**Example:**

Instead of this:



```
return (
  <div>
    <h1>Hello</h1>
    <p>Welcome to React!</p>
  </div>
);

```



**You can use a Fragment like this:**



```
return (
  <>
    <h1>Hello</h1>
    <p>Welcome to React!</p>
  </>
);

```

# **What is a Higher-Order Component (HOC) in React?**

**A Higher-Order Component (HOC) is a function that takes a component and returns a new enhanced component with additional functionality.**

**Why use HOCs?**

To reuse logic across multiple components.

Helps in features like authentication, logging, or data fetching.

**Think of a HOC Like a Wrapper**

Imagine you own a restaurant ðŸ½ï¸. You serve different dishes, but all of them are served on a plate. Instead of making a separate plate for every dish, you use the same plate for all dishes.

The dish is like a React component.
The plate is like a Higher-Order Component (HOC)â€”it wraps the dish and provides a common feature (like serving food).



```
import React, { useEffect } from "react";

// Higher-Order Component (HOC) for logging props
const withLogger = (WrappedComponent) => {
  return (props) => {
    useEffect(() => {
      console.log("Props received:", props);
    }, [props]);

    return <WrappedComponent {...props} />;
  };
};

// Simple component
const UserProfile = ({ name, age }) => (
  <h2>
    User: {name}, Age: {age}
  </h2>
);

// Enhancing UserProfile component with logging HOC
const UserProfileWithLogger = withLogger(UserProfile);

// Using the enhanced component
const App = () => {
  return <UserProfileWithLogger name="John Doe" age={30} />;
};

export default App;

```



**How It Works?**

withLogger is a Higher-Order Component that wraps another component.

Every time the wrapped component receives new props, it logs them to the console.

UserProfileWithLogger is now an enhanced version of UserProfile that logs props whenever they change.
