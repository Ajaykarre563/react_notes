# **What is useState in React?**

useState allows you to store data temporarily while the component is active. However, the data will be lost if you refresh the page because React state is stored only in memory

useState is a React Hook that allows a component to have a state (a value that can change over time). It helps React remember values and update the UI when the value changes.





```
import React,{ useState } from "react"; //hooks are used in components

const Index = () => {
  const arr= (useState(10)); //argument is the initial value of the state is 0
  console.log(arr); //this is the array of the state and function
  console.log(arr[0]); //this is the value of the state means 10
  console.log(arr[1]); //this is the function to update the state means function
  return (
    <div>
    <h1>React Hooks</h1>
    <button>-</button>
    <span>0</span>
    <button>+</button>
    </div>

  );
}

export default Index;

```

```
import React, { useState } from "react"; // Importing React and the useState hook

const Index = () => {
  const [count, setCount] = useState(0); // Corrected syntax for useState

  const incrementCount = () => {
    setCount(count + 1);
  };
  const decrementCount = () => {
    setCount(count - 1);
  };

  return (
    <div>
      <h1>React Hooks</h1>
      <button onClick={decrementCount}>-</button> {/* Fixed incorrect syntax */}
      <span>count : {count} </span>
      <button onClick={incrementCount}>+</button>
    </div>
  );
};

export default Index;

```
