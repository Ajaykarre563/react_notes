# **what is mean by useEffect ?**

**useEffect is a hook in React that lets you perform side effects in your component, like fetching data, subscribing to events, or updating the DOM after a render.**

**In simple terms, it's like saying: "After this component has finished rendering, do this extra work."**

For example, you can use useEffect to log something to the console, fetch data from an API, or update something outside of React's usual render cycle (like interacting with the browser).

You can control when useEffect runs by passing it a second argument:



```
import React, { useEffect, useState } from "react";
// Importing React and two hooks: useEffect (for side effects) and useState (for state management)

const Final = () => {
  // Defining a functional component named 'Final'

  const [count, setCount] = useState(0);
  // Declaring a state variable 'count' with an initial value of 0
  // 'setCount' is the function used to update 'count'

  useEffect(() => {
    console.log("Hello I am coming from useEffect", count);
    // This runs when the component mounts (first renders)
    // It logs a message along with the count value

  });
  // The empty dependency array [] means this effect runs only once when the component mounts

  return (
    <div>
      {/* A container div wrapping the component's content */}
      
      <h1>Learn useEffect hello</h1>
      {/* Displaying a static heading */}

      <h1>{count}</h1>
      {/* Displaying the value of 'count' inside an h1 tag */}

      <button onClick={() => { setCount(count + 1); }}>
        {/* Button that increases 'count' by 1 when clicked */}
        +
      </button>

    </div>
  );
};

export default Final;
// Exporting the 'Final' component so it can be used in other files

```

# **By using Toggle**



```
import React, { useEffect, useState } from "react";
// Importing React and two hooks: useEffect (for side effects) and useState (for state management)

const Final = () => {
  // Defining a functional component named 'Final'

  const [count, setCount] = useState(0);
  // Declaring a state variable 'count' with an initial value of 0
  // 'setCount' is the function used to update 'count'
  const[toggle,setToggle]=useState(true);

  useEffect(() => {
    console.log("Hello I am coming from useEffect", count);
    // This runs when the component mounts (first renders)
    // It logs a message along with the count value

  },[count,toggle]);
  // The empty dependency array [] means this effect runs only once when the component mounts

  return (
    <div>
      {/* A container div wrapping the component's content */}
      
      <h1>Learn useEffect hello</h1>
      {/* Displaying a static heading */}

      <h1 onClick={() => setToggle(!toggle)}>{toggle ? "open" : "close"}</h1>


      <h1>{count}</h1>
      {/* Displaying the value of 'count' inside an h1 tag */}

      <button onClick={() => { setCount(count + 1); }}>
        {/* Button that increases 'count' by 1 when clicked */}
        +
      </button>

    </div>
  );
};

export default Final;
// Exporting the 'Final' component so it can be used in other files

```

# **WindowSize**



```
import React, { useState, useEffect, useCallback } from "react";

const Final = () => {
  const [count, setCount] = useState(0);
  const [toggle, setToggle] = useState(true);
  const [pageWidth, setPageWidth] = useState(window.innerWidth);

  // Function to update page width on resize
  const handleResize = useCallback(() => {
    setPageWidth(window.innerWidth);
  }, []);

  useEffect(() => {
    window.addEventListener("resize", handleResize);

    return () => {
      window.removeEventListener("resize", handleResize);
    };
  }, [handleResize]);

  return (
    <div>
      <h1>Learn useEffect</h1>
      {/* Toggle button */}
      <h1 onClick={() => setToggle(!toggle)}>{toggle ? "open" : "close"}</h1>
      {/* Display counter */}
      <h1>{count}</h1>
      {/* Button to increase counter */}
      <button onClick={() => setCount(count + 1)}>+</button>
      {/* Display page width */}
      <h2>Page Width: {pageWidth}px</h2>
    </div>
  );
};

export default Final;

```
