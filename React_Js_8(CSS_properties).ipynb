# **Complete CSS Properties Guide**

# **1. Text & Font Properties**

color: Sets text color. Example: color: blue;

font-size: Sets text size. Example: font-size: 20px;

font-family: Defines font type. Example: font-family: Arial, sans-serif;

font-weight: Controls boldness. Example: font-weight: bold;

font-style: Sets text style. Example: font-style: italic;

text-align: Aligns text. Example: text-align: center;

text-decoration: Adds underline, overline. Example: text-decoration: underline;

letter-spacing: Sets space between letters. Example: letter-spacing: 2px;

word-spacing: Sets space between words. Example: word-spacing: 5px;

line-height: Controls space between lines. Example: line-height: 1.5;

white-space: Defines whitespace handling. Example: white-space: nowrap;

text-transform: Changes text case. Example: text-transform: uppercase;

# **2. Background & Box Properties**

background-color: Sets background color. Example: background-color: yellow;

background-image: Adds a background image. Example: background-image: url('image.jpg');

background-size: Defines background size. Example: background-size: cover;

background-position: Positions background image. Example: background-position: center;

background-repeat: Controls background repeat. Example: background-repeat: no-repeat;

width & height: Sets element dimensions. Example: width: 200px; height: 100px;

max-width & min-width: Defines max/min width. Example: max-width: 500px;

max-height & min-height: Defines max/min height. Example: min-height: 100px;

# **3. Margin, Padding, & Border**

margin: Creates space outside an element. Example: margin: 10px;

padding: Creates space inside an element. Example: padding: 20px;

border: Sets element border. Example: border: 2px solid black;

border-radius: Rounds element corners. Example: border-radius: 10px;

# **4. Display & Visibility**

display: Defines element display. Example: display: none;

visibility: Controls element visibility. Example: visibility: hidden;

overflow: Handles content overflow. Example: overflow: scroll;

# **5. Positioning & Flexbox**

position: Sets element position. Example: position: absolute;

top / bottom / left / right: Moves elements. Example: top: 10px; left: 20px;

z-index: Defines stack order. Example: z-index: 10;

float: Moves element left/right. Example: float: right;

display: flex: Enables flexbox layout. Example: display: flex; justify-content: center;

# **6. Animation & Effects**

opacity: Controls transparency. Example: opacity: 0.5;

box-shadow: Adds shadow. Example: box-shadow: 5px 5px 10px gray;

transition: Adds animation. Example: transition: all 0.3s ease-in-out;

transform: Rotates, scales, moves. Example: transform: rotate(45deg);

# **7. Cursor & Pointer Events**

cursor: Changes cursor type. Example: cursor: pointer;

pointer-events: Controls click actions. Example: pointer-events: none;

# **What are Props in React?**

Props (short for "Properties") are like arguments that you pass to a function.
They allow you to send data from one component to another.



```
function Greet(props) {
  return <h1>Hello, {props.name}!</h1>;
}

```

# **App.js**



```
import React from "react";
import "./index.css";

let title = "Squid Game"; // Movie title
let image = "/sqi.jpg"; // Movie image
let description = "Squid Game is a thrilling survival drama where contestants play deadly childhood games for a huge cash prize."; // Movie description

// MovieCard component displays an individual movie card
const MovieCard = ({ image, title, description }) => {
  return (
    <div className="movie-card">
      <img src={image} alt={title} className="movie-image" />
      <h2 className="movie-title">{title}</h2>
      <p className="movie-description">{description}</p>
    </div>
  );
};

// App component renders multiple movie cards
const App = () => {
  return (
    <section className="movie-container">
      <MovieCard image={image} title={title} description={description} />
      <MovieCard image={image} title={title} description={description} />
      <MovieCard image={image} title={title} description={description} />
      <MovieCard image={image} title={title} description={description} />
      <MovieCard image={image} title={title} description={description} />
      <MovieCard image={image} title={title} description={description} />
    </section>
  );
};


//function getData(first,last) {
  //console.log(first)
  //console.log(last)
//}
//getData("ajay","karre")
export default App;

```

```
.movie-container {
    display: grid;
    grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
    gap: 20px;
    padding: 20px;
  }
  
  .movie-card {
    background: white;
    border-radius: 8px;
    box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2);
    padding: 15px;
    text-align: center;
  }
  
  .movie-image {
    width: 100%; /* This will stretch the image to the full width of its container */
    height: 400px; /* Increase the height as desired (can be more than 400px too) */
    max-width: 100%; /* Prevents the image from overflowing */
    object-fit: cover; /* Ensures the image maintains its aspect ratio without distortion */
    border-radius: 8px;
  }
  
  
  .movie-title {
    font-size: 20px;
    color: green;
    margin-top: 10px;
  }
  
  .movie-description {
    font-size: 14px;
    color: gray;
  }
  
```
