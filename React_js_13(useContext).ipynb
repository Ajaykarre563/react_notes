**What is useContext in Simple Terms?**

ðŸ”¹ useContext is a React Hook that allows components to access shared data without passing props manually through every level.

ðŸ”¹ It is used with React Context API to share data (like user info, theme, language, etc.) across multiple components.



```
import React, { useState, createContext, useContext } from "react";

// Creating a context to share user details
const UserContext = createContext();

const ParentComponent = () => {
  // useState should be correctly defined
  const [userDetails, setUserDetails] = useState({
    first: "Ajay",
    last: "Karre",
  });

  return (
    // Providing the userDetails to child components using Context
    <UserContext.Provider value={userDetails}>
      <div>
        <h1>useContext Parent Component</h1>
        <ChildComponent />
      </div>
    </UserContext.Provider>  //is used to provide a value (data) to all child components that consume the context.
  );
};

const ChildComponent = () => {
  // Using useContext to get user details
  const userDetails = useContext(UserContext);

  return (
    <div>
      <h2>
        Child Component - User: {userDetails.first} {userDetails.last}
      </h2>
      <SubChildComponent />
    </div>
  );
};

const SubChildComponent = () => {
  // Using useContext in the sub-child component
  const userDetails = useContext(UserContext);

  return (
    <div>
      <h3>Sub Child Component - User: {userDetails.first} {userDetails.last}</h3>
    </div>
  );
};

export default ParentComponent;

```
