## **What is JSX?**

**JSX (JavaScript XML) is a syntax extension for JavaScript that looks like HTML but works inside JavaScript code. It allows you to write HTML-like code in React, making it easier to create UI components.**

**JSX Rules (Simple Explanation)**

Must return a single parent element



```
return (
    <div>
        <h1>Hello</h1>
        <p>Welcome!</p>
    </div>
);

```



**JSX expressions must be wrapped inside {}**



```
const name = "Ajay";
return <h1>Hello, {name}!</h1>; // âœ… Correct

```



**Self-closing tags for empty elements**



```
<img src="image.jpg" alt="Example" />

```



Comments in JSX use {/* ... */}



```
return (
    <div>
        {/* This is a JSX comment */}
        <h1>Hello</h1>
    </div>
);

```

# **How to Use class in JSX?**

**In JSX, you cannot use class (because class is a reserved keyword in JavaScript). Instead, you must use className to apply CSS classes.**



```
import React from 'react';
import './styles.css'; // Import CSS file

function App() {
    return (
        <section className="container">
            <h1 className="heading">Welcome to My Website</h1>
            <p className="description">This is a section with a class.</p>
        </section>
    );
}

export default App;

```



# **What Are Attributes in JSX?**

**In JSX, attributes are properties or values assigned to HTML-like elements to customize or define their behavior. These attributes are similar to HTML attributes, but there are some differences in naming conventions in JSX.**


**Common JSX Attributes**

className (instead of class) for CSS classes.

htmlFor (instead of for) for labeling form elements.

style for inline CSS styles.

onClick for event handling.

And many others, depending on the type of element.



```
import React from 'react';

function App() {
    const containerStyle = {
        padding: '20px',
        backgroundColor: 'lightblue'
    };

    const handleClick = () => {
        alert("Button clicked!");
    };

    return (
        <div className="container" style={containerStyle}>
            <h1>Welcome to My Website</h1>
            <p>This is a paragraph with some inline styles and events.</p>
            <button onClick={handleClick}>Click Me</button>
        </div>
    );
}

export default App;

```



**Explanation:**

**className:**

Used to assign a CSS class to the element. JSX uses className instead of class (to avoid conflict with JavaScript's class keyword).
Example: div className="container"

**style:**

Used to apply inline styles in JSX. The value of style is an object with key-value pairs representing CSS properties.
