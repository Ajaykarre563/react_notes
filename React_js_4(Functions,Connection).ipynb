# **Normal Function vs Arrow Function in Simple Terms**

**Both normal and arrow functions are used in JavaScript, but they have some differences.**

**Normal Function (Traditional Function)**

Uses the function keyword.

Supports this keyword (useful in classes).

Can be called before declaration (hoisting applies).



```
function greet(name) {
  return "Hello, " + name;
}

console.log(greet("Ajay")); // Output: Hello, Ajay

```



✅ Best for object methods and when this is needed.

✅ Can be hoisted (used before declaration).

# **2. Arrow Function (Modern & Shorter)**

Introduced in ES6.

Does not have its own this (inherits from parent scope).

More concise syntax.



```
const greet = (name) => {
  return "Hello, " + name;
};

console.log(greet("Ajay")); // Output: Hello, Ajay

```

# **Summary**

# **index.html (Entry Point - Public Folder)**

This file provides the root HTML structure.

It contains a <div id="root"></div>, where the entire React app is injected.



```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>React App</title>
</head>
<body>
    <div id="root"></div> <!-- React App is injected here -->
</body>
</html>

```



**Key Role: Provides an empty HTML container where React renders its components.**

# **index.js (Starting Point - src/index.js)**

This file is responsible for injecting React components into the index.html file.

It selects div#root and renders the main App component inside it.



```
import React from "react";
import ReactDOM from "react-dom/client";
import App from "./App"; // Importing the main App component

// Selecting the root element in index.html and rendering the App component
const root = ReactDOM.createRoot(document.getElementById("root"));
root.render(<App />);

```



**Key Role: Renders the React app inside index.html by selecting #root.**

# **App.js (Main Component - src/App.js)**

This is the root component of the React app.

It contains the main UI and other components.



```
import React from "react";

function App() {
  return (
    <div>
      <h1>Welcome to My React App</h1>
      <p>This is the main component.</p>
    </div>
  );
}

export default App;

```



📌**Key Role: Defines the main UI and acts as the parent for other components.**

#**How They Work Together**

Browser loads index.html →  id="root"

index.js finds #root and injects <App /> into it.

App.js renders the UI inside #root.

The app is now visible in the browser.



```
index.html  →  index.js  →  App.js  →  Components

```



**Summary:**

index.html → Provides the base HTML structure.

index.js → Connects React to index.html and renders App.js.

App.js → The main React component that displays content.
